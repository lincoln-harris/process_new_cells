```python
# might be easier if we do this
coverage_run_old_trim.index = coverage_run_old_trim['sample']
coverage_run_old_trim = coverage_run_old_trim.drop(columns=['sample'])

coverage_run_new.index = coverage_run_new['sample']
coverage_run_new = coverage_run_new.drop(columns=['sample'])

coverage_run_old_trim
```

```python
# actually think it might be best to make two separate dfs
all_cells = set(list(coverage_run_old_trim.index) + list(coverage_run_new.index))
all_muts = set(list(coverage_run_old_trim.columns) + list(coverage_run_new.columns))

coverage_all_total = pd.DataFrame(index=all_cells, columns=all_muts)
coverage_all_variant = pd.DataFrame(index=all_cells, columns=all_muts)

coverage_all_total = coverage_all_total.fillna(0.0)
coverage_all_variant = coverage_all_variant.fillna(0.0)
```

```python
for mut in list(coverage_all_total.columns):
    for cell in list(coverage_all_total.index):
        
        if cell in list(coverage_run_old_trim.index):
            if mut in list(coverage_run_old_trim.columns):   
                curr_str = coverage_run_old_trim.loc[cell][mut]
            else:                                                # just means this cell didnt have that mut
                curr_str = '0:0'
            v_count = int(curr_str.split(':')[0])
            t_count = int(curr_str.split(':')[1])
            
        elif cell in list(coverage_run_new.index):
            if mut in list(coverage_run_new.columns):
                curr_str = coverage_run_new.loc[cell][mut]
            else:
                curr_str = '0:0'
            v_count = int(curr_str.split(':')[0])
            t_count = int(curr_str.split(':')[1])
        
        coverage_all_total.loc[cell][mut] = t_count    # and now we just fill in 
        coverage_all_variant.loc[cell][mut] = v_count  #   both of them 
```

```python
# ok now i think i want two of these, one for total reads and one for variants
muts_x_sample_df_total = muts_x_sample_df
muts_x_sample_df_variant = muts_x_sample_df.copy() # i think this should do the trick 
```

```python
# why dont we subset these guys for only the cells in our new non-immune lists
coverage_all_total = coverage_all_total[coverage_all_total.index.isin(nonimmune_df.index)]
coverage_all_variant = coverage_all_variant[coverage_all_variant.index.isin(nonimmune_df.index)]
```

```python
# update muts_x_sample_df_TOTAL! 
for cell in list(coverage_all_total.index):
    sample = nonimmune_df.loc[cell].sample_name
    for mut in list(coverage_all_total.columns):
        curr_x_cell = coverage_all_total.loc[cell][mut]
        curr_x_sample = muts_x_sample_df_total.loc[sample][mut]
        
        new_x_sample = curr_x_sample + curr_x_cell 
        muts_x_sample_df_total.loc[sample][mut] = new_x_sample
```

```python
# and now do the same for muts_x_sample_df_VARIANT 
for cell in list(coverage_all_variant.index):
    sample = nonimmune_df.loc[cell].sample_name
    for mut in list(coverage_all_variant.columns):
        curr_x_cell = coverage_all_variant.loc[cell][mut]
        curr_x_sample = muts_x_sample_df_variant.loc[sample][mut]
        
        new_x_sample = curr_x_sample + curr_x_cell 
        muts_x_sample_df_variant.loc[sample][mut] = new_x_sample
```

```python
# ayyyyy
muts_x_sample_df_ratios = muts_x_sample_df_variant / muts_x_sample_df_total
muts_x_sample_df_ratios = muts_x_sample_df_ratios.fillna(0.0)
muts_x_sample_df_ratios = muts_x_sample_df_ratios.drop(columns=['sample'])
muts_x_sample_df_ratios
```
