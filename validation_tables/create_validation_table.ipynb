```python
# lets start with the tumor cell only case
for cell in list(tumor_matches_df.index):
    if cell not in list(valtable_52219.cell):
        print(cell)
        sample = tumor_matches_df.loc[cell].sample_name
        meta_row = supp_meta[supp_meta['Sample Name'] == sample]
        patient_id = list(meta_row['Patient ID'])[0]
        driver_mut_str = list(meta_row['Oncogenic Driver Mutation'])[0]

        try:
            driver_gene = driver_mut_str.split(' ')[0]
            driver_mut = driver_mut_str.split(' ')[1]
        except IndexError:
            driver_gene = driver_mut_str.split(' ')[0]
            driver_mut = ''

        new_row = {} # use a dict instead of a df? 
        new_row['cell'] = cell
        new_row['patient'] = patient_id
        new_row['clinical_driver_gene'] = driver_gene
        new_row['clinical_mutation'] = driver_mut
        new_row['mutations_found'] = ''
        new_row['coverage_to_ROI'] = 0.0
        new_row['clinical_mutation_found_bool'] = 0
        new_row['tumorCell_bool'] = 1
        new_row['sample_name'] = sample

        valtable_52219 = valtable_52219.append(new_row, ignore_index=True)
```

```python
# fuck it let me add the nontumor as well 
for cell in list(nontumor_matches_df.index):
    if cell not in list(valtable_52219.cell):
        print(cell)
        sample = nontumor_matches_df.loc[cell].sample_name
        meta_row = supp_meta[supp_meta['Sample Name'] == sample]
        patient_id = list(meta_row['Patient ID'])[0]
        driver_mut_str = list(meta_row['Oncogenic Driver Mutation'])[0]

        try:
            driver_gene = driver_mut_str.split(' ')[0]
            driver_mut = driver_mut_str.split(' ')[1]
        except IndexError:   # one of these guys is only annotated as 'EGFR' for some reason
            driver_gene = driver_mut_str.split(' ')[0]
            driver_mut = ''

        new_row = {} # use a dict instead of a df? 
        new_row['cell'] = cell
        new_row['patient'] = patient_id
        new_row['clinical_driver_gene'] = driver_gene
        new_row['clinical_mutation'] = driver_mut
        new_row['mutations_found'] = ''
        new_row['coverage_to_ROI'] = 0.0
        new_row['clinical_mutation_found_bool'] = 0
        new_row['tumorCell_bool'] = 0
        new_row['sample_name'] = sample

        valtable_52219 = valtable_52219.append(new_row, ignore_index=True)
```

```python
# fill in 'mutations_found'
valtable_52219.index = valtable_52219.cell   # this will make things easier downstream

for cell in list(cereb_out_new.index):   # outer loop 
    if cell in list(valtable_52219.index): # i dont know why it wouldnt be? 
        curr_muts = ''
        mut_count = 0 
        
        for mut in cereb_out_new.columns:   # inner loop 
            curr = cereb_out_new.loc[cell][mut]
            if curr != '0:0':
                if '*' not in mut: # dont like this 
                    mut_str_edit = mut.split('_')[0] + ' ' + mut.split('_')[1]
                    curr_muts = curr_muts + mut_str_edit + ', '
                    mut_count += 1
       
        if mut_count == 1:   # uggghhhh this is messy 
            curr_muts = curr_muts.replace(',', '') 
            valtable_52219.loc[cell, 'mutations_found'] = curr_muts 
        elif mut_count > 1:
            curr_muts_strip = curr_muts.rstrip(',').rstrip(' ')
            curr_muts_strip = curr_muts_strip.rstrip(',').rstrip(' ')
    
            valtable_52219.loc[cell, 'mutations_found'] = curr_muts_strip 
```

```python
# ok now lets get coverage_to_ROI
for cell in list(cereb_out_new.index): # starting here bc we only want to do the new ones 
    if cell in list(valtable_52219.index): # i dont know why it wouldnt be? 
        
        # lets pick up its meta info 
        meta_line = valtable_52219.loc[cell]
        driver_mut = meta_line.clinical_driver_gene + '_' + meta_line.clinical_mutation
        
        if driver_mut != 'EGFR_del19':   # bc we didnt find any of these, right? 
            try:
                curr_ratio = cereb_out_new.loc[cell, driver_mut]
                if curr_ratio != '0:0':
                    total_reads = int(curr_ratio.split(':')[1])
                    valtable_52219.loc[cell, 'coverage_to_ROI'] = total_reads
                    
            except KeyError:
                #print('mut not in cereb_out_new')
                print(driver_mut)
```
