```python
# the problem with this analysis is that a given cell doesnt have ALL of these variants. 
    # cerebra is only saying these are all of the POTENTIAL amino acid level variants in a given cell
    # so this is probably not the best way of going about it 
def create_mutations_dict(gene):
    '''want to make a dict with ALL of the mutations for a given gene '''
    d = {}
    for cell in ratios_df_sub.index:
        curr_muts = ratios_df_sub.loc[cell][gene]
        if len(curr_muts) > 2: # has mutation -- why are these strs??
            ensps = curr_muts.split(',')

            for ensp in ensps:
                if '?' not in ensp and '=' not in ensp:
                    mut_str = ensp.split('(')[1].split(')')[0]

                    if cell in d:
                        curr = d[cell]
                        curr.append(gene + ' ' + mut_str)
                        d[cell] = curr
                    else:
                        d[cell] = [gene + ' ' + mut_str]
                        
    return(d)
```

```python
egfr_muts_x_cell_d = create_mutations_dict('EGFR')
kras_muts_x_cell_d = create_mutations_dict('KRAS')
braf_muts_x_cell_d = create_mutations_dict('BRAF')
```
