```python
# lets make a brand new df that will hold RATIOS rather than strings
df = pd.DataFrame(index=new_cells_cereb_nonsyno.index, columns=new_cells_cereb_nonsyno.columns)
df = df.fillna(0.000) # i think this will init everything to float? 
df
```

```python
# nice
type(df.loc['A10_B001544'].BRAF_S113I)
```

```python
# now lets fill it in 
for col in list(new_cells_cereb_nonsyno.columns):
    for cell in new_cells_cereb_nonsyno.index:
        curr_str = new_cells_cereb_nonsyno.loc[cell][col]

        v_reads = float(curr_str.split(':')[0])
        t_reads = float(curr_str.split(':')[1])
        
        if t_reads != 0:
            ratio = float(v_reads / t_reads)
        else:
            ratio = 0.0
            
        df.loc[cell][col] = ratio # assignment 
```
