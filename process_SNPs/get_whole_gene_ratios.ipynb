```python
# doin some cool shit here, buh!
b = pd.DataFrame(index=ratios_df_cosmic.index, columns=ratios_df_cosmic.columns) # init blank df 

for cell in ratios_df_cosmic.index:   # outer loop -- row wise
    for gene in list(ratios_df_cosmic.columns):   # inner loop -- column wise
        entry = ratios_df_cosmic.loc[cell][gene]
        entry_ratios = []
        
        if len(entry) > 2: # if entry is not blank
            for x in entry.split(' '):
                ensp = x.split(',')[0]
                ratio = x.split(',')[1]
                
                r = ratio.strip("]") # will this clean them up? 
                r = r.strip("'")
                entry_ratios.append(r)
                
        entry_ratios = set(entry_ratios) # nice!
        
        if entry_ratios:
            b.loc[cell][gene] = entry_ratios # update blank df
```

```python
# nice! 
b
```

```python
b = b.drop('RET', axis=1) # why do I have to do this?!?
b = b.fillna(0)
b
```

```python
import numpy as np
z = pd.DataFrame(index=ratios_df_cosmic.index, columns=ratios_df_cosmic.columns) # think we need another blank df
z = z.fillna(float(0))

for cell in b.index:   # outer loop -- row wise
    for gene in list(b.columns):   # inner loop -- column wise
        
        entry = b.loc[cell][gene]
        v_count = 0
        wt_count = 0
        
        if entry != 0:
            for r in entry:
                v = int(r.split(':')[0].strip('[').strip(']'))
                wt = int(r.split(':')[1].strip('[').strip(']'))
                v_count += v
                wt_count += wt
        
        total_reads = v_count + wt_count
        try:
            new_ratio = float(v_count) / float(total_reads)
        except ZeroDivisionError:
            new_ratio = 0
        
        z.loc[cell][gene] = new_ratio # update
```

```python
# take out rows that are zero across the board
zeros = []
for cell in z.index:
    row = z.loc[cell]
    if (row == 0).all():
        #print(cell)
        zeros.append(cell)
```
