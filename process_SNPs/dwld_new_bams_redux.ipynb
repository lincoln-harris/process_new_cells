```python
seqbot_f_names = list(seqbot_s3_df.file)
seqbot_f_names
```

```python
seqbot_bams_list = []
for f in seqbot_f_names:
    if '.Aligned.out.sorted.bam' in f and '.bai' not in f:
        #print(f)
        seqbot_bams_list.append(f)
len(seqbot_bams_list)
```

```python
# lets figure out the missing ones again
    # old cells + the new (neo osi?) ones
new_nonimmune = pd.read_csv('/Users/lincoln.harris/Desktop/nonimmune_cells_2.18.20.csv', index_col=0)
new_nonimmune
```

```python
# cells I already have vcfs for
! aws s3 ls s3://darmanis-group/singlecell_lungadeno/non_immune/nonImmune_bams_9.27/vcf1/ > out1
existing_vcf_df = pd.read_csv('out1', names=['date', 'time', 'size', 'file'], delim_whitespace=True)
existing_vcf_df   # number seems right
```

```python
existing_vcf_list = [x.strip('.vcf') for x in list(existing_vcf_df.file)]
existing_vcf_list
```

```python
# get the ones we need
still_need = set(new_nonimmune.x) - set(existing_vcf_list)
len(still_need)
```

```python
seqbot_cells_list = [x.split('_')[0] + '_' + x.split('_')[1] for x in seqbot_bams_list]
seqbot_cells_list
```

```python
# this is the list of cells that we need to run that are contained within this seqbot s3 bucket
inter = still_need.intersection(seqbot_cells_list)
len(inter)
```

```python
# what about these remaining ~100? 
remainder = still_need - inter
```

```python
# now go back to the original seqbot s3 dataframe 
seqbot_s3_df
```

```python
# get full s3 paths for these guys 
for f in list(seqbot_s3_df.file):
    if '.Aligned.out.sorted.bam' in f and '.bai' not in f:
        stem = f.split('_')[0] + '_' + f.split('_')[1]
        
        if stem in inter:
            full_path = s3_path_new + f
            cmd = 'aws s3 cp ' + full_path + ' .'
            print(cmd)
```
