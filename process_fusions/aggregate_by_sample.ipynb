```python
# hmmm this only has 3754 cells
sample_cell_key = pd.read_csv('/Users/lincoln.harris/Desktop/laud_process_new_cells/sample_cell_key.csv', index_col=0)
sample_cell_key
```

```python
# lets start by limiting things to these 3754
fusions_x_cell_trim = fusions_x_cell[fusions_x_cell.index.isin(list(sample_cell_key.cell_id))]
fusions_x_cell_trim
```

```python
# lets hunt for just ALK
alk_eml4_cells = []
for cell in fusions_x_cell_trim.index:
    curr_fus = fusions_x_cell_trim.loc[cell].fusions
    
    for elm in curr_fus.split(','):
        if 'EML4--ALK' in elm or 'ALK--EML4' in elm:
            print(cell)
            alk_eml4_cells.append(cell)
```

```python
# and now ROS1
ros1_cd74_cells = []
ros1_ezr_cells = []

for cell in fusions_x_cell_trim.index:
    curr_fus = fusions_x_cell_trim.loc[cell].fusions
    
    for elm in curr_fus.split(','):
        if 'CD74--ROS1' in elm or 'ROS1--CD74' in elm:
            ros1_cd74_cells.append(cell)
        elif 'EZR--ROS1' in elm or 'ROS1--EZR' in elm:
            ros1_ezr_cells.append(cell)
```

```python
ros1_ezr_cells
```

```python
# lets make that by-sample df
x_sample_df = pd.DataFrame(index=set(sample_cell_key.sample_name), columns=['ALK-EML4_fusion', 'ROS1-CD74_fusion', 'ROS1-EZR_fusion'])
x_sample_df = x_sample_df.fillna(0)
x_sample_df
```

```python
# just using index slicing here
for x in alk_eml4_cells:
    meta_row = sample_cell_key[sample_cell_key.cell_id == x]
    curr_sample = list(meta_row.sample_name)[0]
    
    if x_sample_df.loc[curr_sample]['ALK-EML4_fusion'] == 0:
        x_sample_df.loc[curr_sample]['ALK-EML4_fusion'] = 1

        
for x in ros1_cd74_cells:
    meta_row = sample_cell_key[sample_cell_key.cell_id == x]
    curr_sample = list(meta_row.sample_name)[0]
    
    if x_sample_df.loc[curr_sample]['ROS1-CD74_fusion'] == 0:
        x_sample_df.loc[curr_sample]['ROS1-CD74_fusion'] = 1
      
    
for x in ros1_ezr_cells:
    meta_row = sample_cell_key[sample_cell_key.cell_id == x]
    curr_sample = list(meta_row.sample_name)[0]
    
    if x_sample_df.loc[curr_sample]['ROS1-EZR_fusion'] == 0:
        x_sample_df.loc[curr_sample]['ROS1-EZR_fusion'] = 1
```
