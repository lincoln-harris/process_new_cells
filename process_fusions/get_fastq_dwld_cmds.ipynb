```python
import pandas as pd
```

```python
star_out_path = 's3://darmanis-group/singlecell_lungadeno/non_immune/nonImmune_fastqs_9.27/STAR-fus_out/'
```

```python
! aws s3 ls $star_out_path > out
```

```python
star_out_df = pd.read_csv('out', names=['is_prefix', 'file'], delim_whitespace=True)
star_out_df
```

```python
star_out_l = list(star_out_df.file)
star_out_l = [x.strip('/') for x in star_out_l]
star_out_l
```

```python
len(star_out_l)
```

```python
# figure out which cells we still need to generate fusions for
non_immune_cells = pd.read_csv('/Users/lincoln.harris/Desktop/nonimmune_cells_2.18.20.csv', index_col=0)
non_immune_cells
```

```python
inter = set(non_immune_cells.x).intersection(set(star_out_l))
len(inter)
```

```python
still_need = set(non_immune_cells.x) - inter
len(still_need)
```

```python
still_need_df = pd.DataFrame(still_need)
still_need_df.to_csv('fusions_still_needed.csv', index=False)
```

```python
#/////////////////////////////////////////////////////////////////////////////////////
#/////////////////////////////////////////////////////////////////////////////////////
#/////////////////////////////////////////////////////////////////////////////////////
```

```python
# lets get cmds for dwlding them
s3_path_redux = 's3://czb-seqbot/fastqs/190520_A00111_0314_AHJY7FDMXX/rawdata/'
```

```python
! aws s3 ls $s3_path_redux > out
s3_folder_df = pd.read_csv('out', names=['is_prefix', 'folder'], delim_whitespace=True)
s3_folder_df
```

```python
found = []

for parent in list(s3_folder_df.folder):
    stem = parent.split('_')[0] + '_' + parent.split('_')[1]
    full_s3_path = s3_path_redux + parent
    
    if stem in still_need:
        cmd = 'aws s3 cp ' + full_s3_path + ' . --recursive'
        print(cmd)
        
        if stem not in found:
            found.append(stem)
```

```python
# so we only got like a third of them? 
print(len(still_need))
print(len(found))
```

```python
# the rest must be in the rawdata s3 folder 
s3_path_orig = 's3://darmanis-group/singlecell_lungadeno/rawdata/fastq_all/rename/'

! aws s3 ls $s3_path_orig > out
s3_folder_df_orig = pd.read_csv('out', names=['date', 'time', 'size', 'name'], delim_whitespace=True)
s3_folder_df_orig
```

```python
s3_folder_df_orig = s3_folder_df_orig.drop([0])
```

```python
# repeat this process 
for parent in list(s3_folder_df_orig.name):
    stem = parent.split('_')[0] + '_' + parent.split('_')[1]
    full_s3_path = s3_path_orig + parent
    
    if stem in still_need:
        cmd = 'aws s3 cp ' + full_s3_path + ' .'
        print(cmd)
        
        if stem not in found:
            found.append(stem)
```
