```python
import pandas as pd
import os
```

```python
s3_dir = 's3://darmanis-group/lincoln_scratch/LAUD_revisions/StarFusionOut_3.16.20/'
```

```python
! aws s3 ls $s3_dir > out
s3_df = pd.read_csv('out', names=['date', 'time', 'size', 'name'], delim_whitespace=True)
s3_df
```

```python
fusions_still_needed = pd.read_csv('fusions_still_needed.csv')
fusions_still_needed
```

```python
have = [x.split('_')[0] + '_' + x.split('_')[1] for x in list(s3_df.name)]
have
```

```python
# nice
still_need = set(fusions_still_needed['0']) - (set(have))
len(still_need)
```

```python
#/////////////////////////////////////////////////////////////////////////////////////
#/////////////////////////////////////////////////////////////////////////////////////
#/////////////////////////////////////////////////////////////////////////////////////
```

```python
# now lets try and gen s3 cp cmds
s3_path_redux = 's3://czb-seqbot/fastqs/190520_A00111_0314_AHJY7FDMXX/rawdata/'

! aws s3 ls $s3_path_redux > out
s3_folder_df = pd.read_csv('out', names=['is_prefix', 'folder'], delim_whitespace=True)
s3_folder_df

found = []

for parent in list(s3_folder_df.folder):
    stem = parent.split('_')[0] + '_' + parent.split('_')[1]
    full_s3_path = s3_path_redux + parent
    
    if stem in still_need:
        cmd = 'aws s3 cp ' + full_s3_path + ' . --recursive'
        print(cmd)
        
        if stem not in found:
            found.append(stem)
```

```python
# alright so 97 from redux bucket
len(found)
```

```python
# the rest must be in the rawdata s3 folder 
s3_path_orig = 's3://darmanis-group/singlecell_lungadeno/rawdata/fastq_all/rename/'

! aws s3 ls $s3_path_orig > out
s3_folder_df_orig = pd.read_csv('out', names=['date', 'time', 'size', 'name'], delim_whitespace=True)

s3_folder_df_orig = s3_folder_df_orig.drop([0])

# repeat this process 
for parent in list(s3_folder_df_orig.name):
    stem = parent.split('_')[0] + '_' + parent.split('_')[1]
    full_s3_path = s3_path_orig + parent
    
    if stem in still_need:
        cmd = 'aws s3 cp ' + full_s3_path + ' .'
        print(cmd)
        
        if stem not in found:
            found.append(stem)
```
