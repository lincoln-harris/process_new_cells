```python
import pandas as pd
```

```python
cereb_vcf = pd.read_csv('/Users/lincoln.harris/Desktop/cerebra_out_all_nonimmune.csv', index_col=0)
cereb_vcf
```

```python
cereb_gvcf = pd.read_csv('tumor_cell_gvcf_run.csv', index_col=0)
cereb_gvcf
```

```python
# subset to just non-syno muts
keep_cols_vcf = []
for x in list(cereb_vcf.columns):
    mutstr = x.split('_')[1]
    if mutstr[0] != mutstr[-1]:
        keep_cols_vcf.append(x)
        
# lets get rid of non-syno muts -- for gvcf df
keep_cols_gvcf = []
for x in list(cereb_gvcf.columns):
    mutstr = x.split('_')[1]
    if mutstr[0] != mutstr[-1]:
        keep_cols_gvcf.append(x)
        
cereb_vcf = cereb_vcf[keep_cols_vcf]
cereb_gvcf = cereb_gvcf[keep_cols_gvcf]
```

```python
# now lets subset to just tumor cells 
tumor_cells_df = pd.read_csv('/Users/lincoln.harris/Desktop/laud_process_new_cells/sample_cell_key.csv', index_col=0)
tumor_cells_list = list(tumor_cells_df.cell_id)
```

```python
cereb_vcf = cereb_vcf[cereb_vcf.index.isin(tumor_cells_list)]
cereb_gvcf = cereb_gvcf[cereb_gvcf.index.isin(tumor_cells_list)]
```

```python
print(cereb_vcf.shape)
print(cereb_gvcf.shape)
```

```python
# ok now i want two tables, one for total reads and one for variant reads 
all_rows = set(list(cereb_vcf.index) + list(cereb_gvcf.index))
all_cols = set(list(cereb_vcf.columns) + list(cereb_gvcf.columns))
big_df = pd.DataFrame(index=all_rows, columns=all_cols)
big_df = big_df.fillna(0)
big_df
```

```python
# make two of 'em
big_df_v = big_df
big_df_t = big_df.copy()
```

```python
# lets start with the variants df
for cell in list(big_df_v.index):
    for mut in list(big_df_v.columns):
        
        if cell in list(cereb_vcf.index) and mut in list(cereb_vcf.columns): # in the vcf df
            curr = cereb_vcf.loc[cell][mut]
            v_count = int(curr.split(':')[0])
            big_df_v.loc[cell][mut] = v_count
```

```python
# and now do the total counts df
for cell in list(big_df_t.index):
    for mut in list(big_df_t.columns):
        
        if cell in list(cereb_gvcf.index) and mut in list(cereb_gvcf.columns): # in the gvcf df 
            # do something
            curr = cereb_gvcf.loc[cell][mut]
            t_count = int(curr.split(':')[1])
            big_df_t.loc[cell][mut] = t_count
```

```python
# now what? -- break it down by sample 
by_sample_v = pd.DataFrame(index = set(tumor_cells_df.sample_name), columns = big_df_t.columns)
by_sample_t = pd.DataFrame(index = set(tumor_cells_df.sample_name), columns = big_df_t.columns)
by_sample_v = by_sample_v.fillna(0)
by_sample_t = by_sample_t.fillna(0)
```

```python
# first lets fill in by_sample_v
for cell in list(big_df_v.index):
    curr_sample = tumor_cells_df.loc[cell]['sample_name']
    for mut in list(big_df_v.columns):

        curr_x_cell = big_df_v.loc[cell][mut]
        curr_x_sample = by_sample_v.loc[curr_sample][mut]
        curr_x_sample = curr_x_sample + curr_x_cell
        
        by_sample_v.loc[curr_sample][mut] = curr_x_sample
```

```python
# now lets fill in by_sample_t
for cell in list(big_df_t.index):
    curr_sample = tumor_cells_df.loc[cell]['sample_name']
    for mut in list(big_df_t.columns):

        curr_x_cell = big_df_t.loc[cell][mut]
        curr_x_sample = by_sample_t.loc[curr_sample][mut]
        curr_x_sample = curr_x_sample + curr_x_cell
        
        by_sample_t.loc[curr_sample][mut] = curr_x_sample
```

```python
by_sample_t
```

```python
# now lets do the division
    # think this should work!!
by_sample_ratios = pd.DataFrame(index=list(by_sample_v.index), columns=list(by_sample_v.columns))
by_sample_ratios = by_sample_ratios.fillna(0.0) # want these init to floats

for sample in list(by_sample_ratios.index):
    for mut in list(by_sample_ratios.columns):
        v_entry = by_sample_v.loc[sample][mut]
        t_entry = by_sample_t.loc[sample][mut]
        
        if v_entry > 0 and t_entry > 0:   # most simple case
            curr = v_entry / t_entry
            
        elif v_entry == 0 and t_entry > 0:   # positive coverage case
            curr = -1
            
        else:   # zero coverage case
            curr = 0 
            
        by_sample_ratios.loc[sample][mut] = curr   # assign
```
