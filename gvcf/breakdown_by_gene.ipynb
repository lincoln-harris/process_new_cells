```python
''' ok now i want to make a by-gene dataframe
    the big question here is whether to include the driver mutations or not '''
```

```python
import pandas as pd
```

```python
x_sample_v = pd.read_csv('x_sample_v_counts.csv', index_col=0)
x_sample_t = pd.read_csv('x_sample_t_counts.csv', index_col=0)
```

```python
x_sample_v
```

```python
# maybe the easiest way to start is by generating a by-sample list of all the muts
    # start with empty dict
x_sample_d = {}
for sample in list(x_sample_v.index):
    x_sample_d[sample] = []
```

```python
for sample in list(x_sample_v.index):
    curr_muts = []
    for mut in list(x_sample_v.columns):
        curr = x_sample_v.loc[sample][mut]
        if curr !=0 and '*' not in mut:
            curr_muts.append(mut)
            
    x_sample_d[sample] = curr_muts
```

```python
# great! now lets convert this bad boy to a dataframe, then csv
x_sample_d
```

```python
# sort of gonna do this manually here
    # this looks good 
x_sample_df = pd.DataFrame(index=list(x_sample_d.keys()), columns=['mutations'])

# for loop it up!!
for key in list(x_sample_d.keys()):
    x_sample_df.loc[key].mutations = x_sample_d[key]

x_sample_df
```

```python
x_sample_df.to_csv('samples_all_mutations_3.30.20.csv')
```

```python
#////////////////////////////////////////////////////////////////////
#////////////////////////////////////////////////////////////////////
#////////////////////////////////////////////////////////////////////
```

```python
# now lets try and do the whole-gene summarizing, MINUS the driver muts, for now
driver_muts = ["EGFR_L858R", "EGFR_K745_A750>T", "EGFR_E709_T710>D", "KRAS_G12A",
                "KRAS_G12C", "KRAS_G12S", "KRAS_G12F", "KRAS_G12V", "KRAS_G13D",
                "KRAS_Q61H", "KRAS_Q61K", "KRAS_C185S", "BRAF_V600E"]
```

```python
# this feels really wierd
x_sample_v = x_sample_v.iloc[:, ~x_sample_v.columns.isin(driver_muts)]
x_sample_t = x_sample_t.iloc[:, ~x_sample_t.columns.isin(driver_muts)]
```

```python
# now lets make an samples x genes df 
genes_list = set([x.split('_')[0] for x in list(x_sample_t.columns)])
samples_x_genes = pd.DataFrame(index=list(x_sample_v.index), columns=genes_list)
samples_x_genes = samples_x_genes.fillna(0)
samples_x_genes
```

```python
# now we need to make two of these, one for TOTAL and one for VARIANT reads
samples_x_genes_v = samples_x_genes
samples_x_genes_t = samples_x_genes.copy()
```

```python
# now lets fill 'em in -- for VARIANT df
for sample in list(x_sample_v.index):
    for mut in list(x_sample_v.columns):
        
        curr_gene = mut.split('_')[0]
        x_mut_val = x_sample_v.loc[sample][mut]
        x_gene_val = samples_x_genes_v.loc[sample][curr_gene]
        
        x_gene_val_new = x_gene_val + x_mut_val
        samples_x_genes_v.loc[sample][curr_gene] = x_gene_val_new
```

```python
# now lets do the same -- for TOTAL df
for sample in list(x_sample_t.index):
    for mut in list(x_sample_t.columns):
        
        curr_gene = mut.split('_')[0]
        x_mut_val = x_sample_t.loc[sample][mut]
        x_gene_val = samples_x_genes_t.loc[sample][curr_gene]
        
        x_gene_val_new = x_gene_val + x_mut_val
        samples_x_genes_t.loc[sample][curr_gene] = x_gene_val_new
```

```python
sample_x_genes_ratios = samples_x_genes_v / samples_x_genes_t
sample_x_genes_ratios = sample_x_genes_ratios.fillna(0)
sample_x_genes_ratios
```

```python
# this looks good for now
sample_x_genes_ratios.to_csv('samples_x_genes_3.30.20.csv')
```
