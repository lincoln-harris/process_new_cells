```python
import pandas as pd
```

```python
# vcf, all cells
cereb_vcf = pd.read_csv('/Users/lincoln.harris/Desktop/cerebra_out_all_nonimmune.csv', index_col=0)
cereb_vcf
```

```python
cereb_gvcf = pd.read_csv('tumor_cell_gvcf_run.csv', index_col=0)
cereb_gvcf
```

```python
# lets get rid of non-syno muts
keep_cols_vcf = []
for x in list(cereb_vcf.columns):
    mutstr = x.split('_')[1]
    if mutstr[0] != mutstr[-1]:
        #print(mutstr)
        keep_cols_vcf.append(x)
        
# lets get rid of non-syno muts -- for gvcf df
keep_cols_gvcf = []
for x in list(cereb_gvcf.columns):
    mutstr = x.split('_')[1]
    if mutstr[0] != mutstr[-1]:
        #print(mutstr)
        keep_cols_gvcf.append(x)
```

```python
cereb_vcf = cereb_vcf[keep_cols_vcf]
cereb_gvcf = cereb_gvcf[keep_cols_gvcf]
cereb_vcf
```

```python
# im curious how often gvcf picks something up but the vcf doesnt
    # nice
for cell in list(cereb_gvcf.index):
    for mut in list(cereb_gvcf.columns):
        if cell in list(cereb_vcf.index) and mut in list(cereb_vcf.columns):
            curr_gvcf = cereb_gvcf.loc[cell][mut]
            curr_vcf = cereb_vcf.loc[cell][mut]

            if curr_gvcf != curr_vcf:
                print('gvcf: %s' % curr_gvcf)
                print('vcf: %s' % curr_vcf)
                print(' ')
```
